```
#import all Numpy, Seaborn, and Matplotlib libraries 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def conv2D(image2D, kernel3x3):
        conv2D = np.zeros((len(image2D)-2, len(image2D)-2))
        for i in range(len(image2D)-2):
          for j in range(len(image2D)-2):
            conv2D[i][j] = np.sum(image2D[i:i+3, j:j+3]*kernel3x3)
        return conv2D

#load Cat text file
img2D = np.loadtxt('2dcnn.csv', delimiter=',')
sns.heatmap(img2D, cmap='gray')
plt.title('Actual Image ' + str(img2D.shape))
plt.show()

edf3x3 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

conv_img = conv2D(img2D, edf3x3)
sns.heatmap(conv_img, cmap='gray')
plt.title('Convolution After Edge Detection Filter ' + ' - Size = ' + str(conv_img.shape))
plt.show()
```
